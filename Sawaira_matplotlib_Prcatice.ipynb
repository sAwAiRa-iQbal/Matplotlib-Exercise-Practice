# **Anatomy of Matplotlib**

```python
# step 0: import matplotlib and get it ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

#step1: prepare data
x = [1,2,3,4]
y = [11,22,33,44]

#step2: Setup plot
fig, ax = plt.subplots(figsize=(6,4))

#step3: plot data
ax.plot(x,y)

#step4: costumize plot
ax.set(title = "sample plot",
         xlabel = "x-axis",
         ylabel = "y-axis",
         )

# step5: save and show (you save the whole figure)
#fig.savefig("images/sample-plot.png")
plt.show()
```

# which one you should use? (pyplot vs matplotlib 00 method?)

* when plotting something quickly. okay to use the pyplot method

* when plotting something more adddvanced, use the OO Method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

## **pyplot method**

```python
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             figsize=(10,6));
```

# **OO Method**

```python
# OO method mixed with pyplot
fix, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",   #Specifies that the color of the scatter points should be based on the values in the target column (possibly used to differentiate classes or categories).
             ax=ax);
ax.set_xlim([45, 100]);
```

```python
over_50.target.values
```

```python
# Object Oriented method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# plot the data
scatter = ax.scatter(x =over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

# customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol");

# add a legend
ax.legend(*scatter.legend_elements(),title="Target");

# add a horizontal line
ax.axhline(over_50["chol"].mean(),
          color='red',
          linestyle='--');
```

```python
over_50.head()
```

```python
#subplots of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10,6),
                               sharex=True)

# plot the data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"])

# customize the plot
ax0.set(title="Heart Disease and Cholesterol Levels",
        xlabel="Age",
        ylabel="Cholesterol")

#Add legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

# add a horizontal mean line
ax0.axhline(over_50["chol"].mean(),
            color="red",
            linestyle="--")

#add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"])

# customize the plot
ax1.set(title="Heart Disease and Max Heart Rate",
        xlabel="Age",
        ylabel="Max Heart Rate")

#add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")

# add a horizontal mean line
ax1.axhline(over_50["thalach"].mean(),
            color="red",
            linestyle="--")

# ad a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold"); # Changed 'subtitle' to 'suptitle' to set the title of the figure
```

```python
plt.style.use('seaborn-whitegrid')


# Object Oriented method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# plot the data
scatter = ax.scatter(x =over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap ='winter')      #change color scheme

# customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol");

# add a legend
ax.legend(*scatter.legend_elements(),title="Target");

# add a horizontal line
ax.axhline(over_50["chol"].mean(),
          color='red',
          linestyle='--');
```

```python
# customizing x and y axis limitations
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10),
                               sharex=True)
# add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="winter")

# add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"],
                     cmap="winter")

# customize ax0
ax0.set(title="Heart Disease and Cholestrol Levels",
       ylabel="Cholestrol")
ax0.set_xlim([50, 80])

# customize ax1
ax1.set(title="Heart Disease and Max Heart Rate",
       xlabel="Age",
       ylabel="Max Heart Rate")
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])
# add legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target");

# add legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target");

# add horizontal line ax0
ax0.axhline(y=over_50["chol"].mean(),
          linestyle='--');

# add horizontal line ax1
ax1.axhline(y=over_50["thalach"].mean(),
          linestyle='--');

# add title to figure
fig.suptitle("Heart disease Analysis", fontsize=16, fontweight="bold");
```
